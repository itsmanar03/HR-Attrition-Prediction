```python
df = df.drop(['EmployeeCount','Over18','StandardHours', 'EmployeeNumber'],axis=1)
```

```python
cols_encode = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime']
df = pd.get_dummies(df, columns=cols_encode)
```

```python
le = LabelEncoder()
df['Attrition'] = le.fit_transform(df['Attrition'])
```

```python
df['Attrition']
```

```python
x = df.drop(["Attrition"], axis=1)
y = df["Attrition"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
model_SVC = SVC()
model_SVC.fit(x_train, y_train)
```

```python
y_pred_train = model_SVC.predict(x_train)
accuracy_train = accuracy_score(y_train, y_pred_train)

y_pred_test = model_SVC.predict(x_test)
accuracy_test = accuracy_score(y_test, y_pred_test)

print("Train Accuracy: %.2f%%" % (accuracy_train * 100.0))
print("Test Accuracy: %.2f%%" % (accuracy_test * 100.0))
```

```python
confusion_matrix(y_test,y_pred_test)
```

```python
class_report = classification_report(y_test,y_pred_test)
print(class_report)
```

```python
model_LogReg = LogisticRegression()
model_LogReg.fit(x_train, y_train)
```

```python
y_pred_train1 = model_LogReg.predict(x_train)
accuracy_train1 = accuracy_score(y_train, y_pred_train1)

y_pred_test1 = model_LogReg.predict(x_test)
accuracy_test1 = accuracy_score(y_test, y_pred_test1)

print("Train Accuracy: %.2f%%" % (accuracy_train1 * 100.0))
print("Test Accuracy: %.2f%%" % (accuracy_test1 * 100.0))
```

```python
class_report1 = classification_report(y_test,y_pred_test1)
print(class_report1)
```

```python
confusion_matrix(y_test,y_pred_test1)
```

```python
model_LogReg2 = LogisticRegression(class_weight='balanced')
model_LogReg2.fit(x_train,y_train)
```

```python
y_pred_train2 = model_LogReg2.predict(x_train)
accuracy_train2 = accuracy_score(y_train, y_pred_train2)

y_pred_test2 = model_LogReg2.predict(x_test)
accuracy_test2 = accuracy_score(y_test, y_pred_test2)

print("Train Accuracy: %.2f%%" % (accuracy_train2 * 100.0))
print("Test Accuracy: %.2f%%" % (accuracy_test2 * 100.0))
```

```python
class_report2 = classification_report(y_test,y_pred_test2)
print(class_report2)
```

```python
smote = SMOTE()
x_train_smote,y_train_smote = smote.fit_resample(x_train,y_train)
```

```python
model_LogReg3 = LogisticRegression()
model_LogReg3.fit(x_train_smote,y_train_smote)
```

```python
y_pred_train3 = model_LogReg3.predict(x_train)
accuracy_train3 = accuracy_score(y_train, y_pred_train3)

y_pred_test3 = model_LogReg3.predict(x_test)
accuracy_test3 = accuracy_score(y_test, y_pred_test3)

print("Train Accuracy: %.2f%%" % (accuracy_train3 * 100.0))
print("Test Accuracy: %.2f%%" % (accuracy_test3 * 100.0))
```

```python
class_report3 = classification_report(y_test,y_pred_test3)
print(class_report3)
```
